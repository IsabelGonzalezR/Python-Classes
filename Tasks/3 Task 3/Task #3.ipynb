**1. Comparativa entre Rusia y Ecuador** 

```python
# Partidos totales Rusia
partidos_Rusia = partidos_fifa_copa_mundial[
    (partidos_fifa_copa_mundial['equipo_1'] == 'Rusia') | (partidos_fifa_copa_mundial['equipo_2'] == 'Rusia')]
# Partidos totales Ecuador
partidos_Ecuador = partidos_fifa_copa_mundial[
    (partidos_fifa_copa_mundial['equipo_1'] == 'Ecuador') | (partidos_fifa_copa_mundial['equipo_2'] == 'Ecuador')]
#Partidos totales por equipo
partidos_Rusia_total = len(partidos_Rusia)
partidos_Ecuador_total = len(partidos_Ecuador)

print(f'Rusia jugó un total de {partidos_Rusia_total} partidos.')
print(f'Ecuador jugó un total de {partidos_Ecuador_total} partidos.')
```

```python
# Partidos totales en los que Rusia participó
partidos_Rusia_total = partidos_fifa_copa_mundial[(partidos_fifa_copa_mundial['equipo_1'] == 'Rusia') | (partidos_fifa_copa_mundial['equipo_2'] == 'Rusia')]
partidos_Rusia_total= partidos_Rusia_total.sort_values('fecha')
```

```python
# Participación de Rusia en la Copa Mundial
plt.figure(figsize=(10, 5))
plt.plot(partidos_Rusia_total['fecha'], range(1, len(partidos_Rusia_total) + 1), marker='o', linestyle='-', color='purple', label='Rusia')
plt.title('Participación de Rusia en la Copa Mundial')
plt.xlabel('Fechas de los Partidos')
plt.ylabel('Número de Partidos')
plt.legend() 
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('participacion_rusia.png') #Guardando Figura
plt.show()
```

```python
# Partidos totales en los que Ecuador participó
partidos_Ecuador_total = partidos_fifa_copa_mundial[(partidos_fifa_copa_mundial['equipo_1'] == 'Ecuador') | (partidos_fifa_copa_mundial['equipo_2'] == 'Ecuador')]
partidos_Ecuador_total = partidos_Ecuador_total.sort_values('fecha')
```

```python
# Participación de Ecuador en la Copa Mundial
plt.figure(figsize=(12, 5))
plt.plot(partidos_Ecuador_total['fecha'], range(1, len(partidos_Ecuador_total) + 1), marker='o', linestyle='-', color='green', label='Ecuador')
plt.title('Participación de Ecuador en la Copa Mundial')
plt.xlabel('Fechas de los Partidos')
plt.ylabel('Número de Partidos')
plt.legend() 
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('participacion_ecuador.png') #Guardando Figura
plt.show()
```
