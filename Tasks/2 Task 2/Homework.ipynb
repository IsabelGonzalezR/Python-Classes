```python
# Comparar Escalera_vida de Ecuador y promedio global
diferencia_Ecuador = (5.764949101668138 - 5.4371550357397735)
diferencia_Ecuador
```

```python
# Comparar Escalera_vida de Canada y promedio global
diferencia_Canada = (7.423908380361704- 5.4371550357397735)
diferencia_Canada
```

```python
# Diferencia Escalera_vida de Canada y Ecuador
diferencia_dos_paises = (7.423908380361704 -5.764949101668138)
diferencia_dos_paises
```

```python
paises = ["Canadá", "Ecuador"]
diferencias = [diferencia_dos_paises, -diferencia_dos_paises]
fig, ax = plt.subplots()
ax.bar(paises, diferencias, color=['yellow', 'green'])
ax.set_xlabel('Países')
ax.set_ylabel('Diferencia')
plt.title('Diferencia escalera vida')
plt.show()
```

```python
# Comparar la fecilidad en la percepcion de calidad de vida 'escalera_vida'
#Canada tiene un mejor estilo de vida promedio del  mundo,
# tambien superior a Ecuador por eso lo hace un pais mas feliz
# Canada es un pais desarrollado tiene mejores condiciones de vida 
# en el ambito laboral, social, economico.
```

```python
# Comparar afecto positivo Ecuador y Promedio Global
diferencia_Ecuador = (0.8453240577991192 - 0.709368234872818)
diferencia_Ecuador
```

```python
# Comparar afecto positivo Canada y Promedio Global
diferencia_Canada = (0.8558318018913269 - 0.709368234872818 )
diferencia_Canada
```

```python
# Comparar afecto positivo entre Canada y Ecuador
diferencia_afecto = (0.8558318018913269 - 0.8453240577991192)
diferencia_afecto
```

```python
paises = ["Canadá", "Ecuador"]
diferencias = [diferencia_afecto, diferencia_afecto]
fig, ax = plt.subplots()
ax.bar(paises, diferencias, color=['orange', 'blue'])
ax.set_xlabel('Países')
ax.set_ylabel('Diferencia')
plt.title('Diferencia afecto posito')
plt.show()
```

```python
# Canada tiene mayor promedio global, tambien a diferencia de Ecuador
# tiene mayor efecto positivo en la poblacion de vivencias, reuniones
# viajes, risas, gozar la vida, Canada es uno de los mejores paises del mundo.
```
